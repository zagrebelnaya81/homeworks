Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def DevisionAndMultiply(arr):
  mean = np.mean(arr)
  prepared_data = np.where( arr >= mean , arr, arr*2)
  prepared_data = np.where( arr < mean, prepared_data, prepared_data/4)
  return prepared_data
DevisionAndMultiply(data)
```

Застосувати отриману функцію до 3-ї колонки


```python
third_column = data[:,2]
DevisionAndMultiply(third_column)
```
