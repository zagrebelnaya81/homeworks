новая фича, смотрим сколько в винах всего химических примисей(соединяем количество хлоридов, диоксида серы общий, сульфаты)

```python
result['chemical impurities'] = result['total sulfur dioxide']+result['chlorides']+result['sulphates']
print (result['chemical impurities'])

```

```python
# Розділення датасету на навчальну та контрольну вибірки
X =  result.values[:, 0:10]
print(X)
Y =  result.values[:, 11]
print(Y)
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
X_train, X_validation, Y_train, Y_validation = train_test_split(X_scaled, Y, test_size=0.20, random_state=1)
X_train
```

```python
model = sm.OLS(Y_train, X_train)
results = model.fit()
print(results.summary())
```

```python
print("Parameters: ", results.params)
print("R2: ", results.rsquared)
print("Standard errors: ", results.bse)
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, Y_train)
predict = model.predict(X_validation)
score = model.score(X_validation, Y_validation)
print(score)
report = classification_report(Y_validation, predict)
print(report)
```

```python
cm = confusion_matrix(Y_validation, model.predict(X_validation))
cm
```

```python
model = make_pipeline(StandardScaler(),
                    SGDRegressor(max_iter=1000, tol=1e-3))
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_validation)
predictions
```

```python
print(accuracy_score(Y_validation, predictions.round()))
print(confusion_matrix(Y_validation, predictions.round(), normalize=None))
print(classification_report(Y_validation, predictions.round()))
```
